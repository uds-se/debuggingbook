# Release Notes for The Debugging Book

This book comes with version numbers; these correspond to the version numbers in [the Python pip package](Importing.ipynb).

## Version 1.1 (in progress)

* We now require Python 3.9 or later.
* Made an optimization in DeltaDebugger.dd() to distribute changes more evenly over the input space. This will change `dd()` results in some cases, but overall improve performance. Thanks to Olad Chitil and Joanna Sharrad for reporting this!
* Fixed two [issues in DeltaDebugger.dd()](https://github.com/uds-se/debuggingbook/commit/95ab3117372bff459eede743e57cf974a9d022c4) ([#45]((https://github.com/uds-se/debuggingbook/issues/44), [#46](https://github.com/uds-se/debuggingbook/issues/46)) Thanks to @TheSilvus for reporting this!
* Added missing dependencies to the debuggingbook pip package] ([#44]((https://github.com/uds-se/debuggingbook/issues/44))) Thanks to @TheSilvus for reporting this!
* Fixed a warning '.gitignore is a symbolic link' during git checkout  ([#43]((https://github.com/uds-se/debuggingbook/issues/43))) Thanks to @rjc for reporting this!
